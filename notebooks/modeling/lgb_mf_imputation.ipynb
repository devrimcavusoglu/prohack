```python
from lightgbm import LGBMRegressor
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
```

```python
def plot_lgb_history(history):
    for i, key in enumerate(history.evals_result_):
        if i == 0:
            res = pd.DataFrame(history.evals_result_[key])
            res['index'] = np.arange(len(history.evals_result_[key]['l2']))
            res['type'] = key
        temp = pd.DataFrame(history.evals_result_[key])
        temp['type'] = key
        temp['index'] = np.arange(len(history.evals_result_[key]['l2']))
        
        res = res.append(temp, ignore_index=True)
        del temp; gc.collect()
    
    fig = px.scatter(data_frame=res, x='index', y='l2', color='type',
                 title='Learning curve')
    fig.show()
```

```python
train = pd.get_dummies(train)
```

```python
scaler = MinMaxScaler()

train_temp = train.drop(['galactic year', 'y'], axis=1)
train_sc = scaler.fit_transform(train_temp)
train_sc = pd.DataFrame(train_sc, columns=train_temp.columns)
del train_temp; gc.collect()

train_sc['galactic year'] = train['galactic year']
train_sc['y'] = train['y']
train_sc
```

```python
last_n_years = 3
val_year = train['galactic year'].unique()[-last_n_years]
train_set = train[train['galactic year']<val_year]
val_set = train[train['galactic year']>=val_year]
```

```python
y = 'y'
X = train_set.columns.drop(y)
```

```python
X_train, y_train = train_set[X].to_numpy(), train_set[y].to_numpy()
X_val, y_val = val_set[X].to_numpy(), val_set[y].to_numpy()

eval_set = [(X_train, y_train), (X_val, y_val)]
```

```python
# num_leaves << 2^(max_depth)

lgb_reg = LGBMRegressor(
    boosting_type='gbdt',
    num_leaves=8,
    max_depth=4,
    learning_rate=0.1,
    n_estimators=500,
    subsample_for_bin=500,
    objective='regression',
    class_weight=None,
    min_split_gain=0,
    min_child_weight=0.001,
    min_child_samples=10,
    subsample=0.8,
    subsample_freq=30,
    colsample_bytree=0.9,
    colsample_bynode=0.1,
    reg_alpha=0.9,
    reg_lambda=0.9,
    random_state=None,
    n_jobs=-1,
    silent=True,
    importance_type='split'
)

lgbm = lgb_reg.fit(X=X_train, y=y_train,
                  eval_set = eval_set, verbose=0)

y_pred_train = lgbm.predict(train_set[X])
y_pred_val = lgbm.predict(val_set[X])

train_rmse = mean_squared_error(y_true=y_train, y_pred=y_pred_train, squared=False)
val_rmse = mean_squared_error(y_true=y_val, y_pred=y_pred_val, squared=False)
print(f'RMSE Train: {train_rmse: .8f}')
print(f'RMSE Valid: {val_rmse: .8f}\tRatio (val/train): {val_rmse/train_rmse: .5f}')

plot_lgb_history(lgbm)
```

```python
test = pd.get_dummies(test)
```

```python
galaxies_unseen_test = ['galaxy_Andromeda XIX[60]', 'galaxy_NGC 5253', 'galaxy_Tucana Dwarf',
       'galaxy_Andromeda XXIV', 'Population..ages.15a..64..millions.', 'galaxy_Hercules Dwarf',
       'Unemployment..youth....ages.15a..24.', 'galaxy_Andromeda XXII[57]', 'galaxy_Andromeda XVIII[60]',
       'galaxy_Triangulum Galaxy (M33)', 'galaxy_Andromeda XII']

for galaxy in galaxies_unseen_test:
    test[galaxy] = 0
    
X_test = test[X].to_numpy()
```

```python
y_pred_test = lgbm.predict(X_test)
```
