```python
from lightgbm import LGBMRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
```

```python
def plot_lgb_history(history):
    for i, key in enumerate(history.evals_result_):
        if i == 0:
            res = pd.DataFrame(history.evals_result_[key])
            res['index'] = np.arange(len(history.evals_result_[key]['l2']))
            res['type'] = key
        temp = pd.DataFrame(history.evals_result_[key])
        temp['type'] = key
        temp['index'] = np.arange(len(history.evals_result_[key]['l2']))
        
        res = res.append(temp, ignore_index=True)
        del temp; gc.collect()
    
    fig = px.scatter(data_frame=res, x='index', y='l2', color='type',
                 title='Learning curve')
    fig.show()
```

```python
drop_cols = ['galactic year']
```

```python
# galaxies_unseen_test = list(set(train.galaxy.unique()) - set(test.galaxy.unique()))
# train = train[~train.galaxy.isin(galaxies_unseen_test)].copy()

# date = train['galactic year'].min()

# train['year'] = ((train['galactic year'] - date) / 1000 + 1).astype(int)
# test['year'] = ((test['galactic year'] - date) / 1000 + 1).astype(int)

# train.set_index('year', inplace=True)
# test.set_index('year', inplace=True)

train.drop(drop_cols, axis=1, inplace=True)
test.drop(drop_cols, axis=1, inplace=True)
```

```python
train = pd.get_dummies(train)
```

```python
y = 'y'
X = train.columns.drop(y)
```

```python
X_train, X_val, y_train, y_val = train_test_split(train[X].to_numpy(), train[y].to_numpy(),
                                                  test_size=0.2, random_state=1)
X_train.shape, X_val.shape, y_train.shape, y_val.shape
```

```python
eval_set = [(X_train, y_train), (X_val, y_val)]
```

```python
# num_leaves << 2^(max_depth)

lgb_reg = LGBMRegressor(
    boosting_type='gbdt',
    num_leaves=8,
    max_depth=4,
    learning_rate=0.1,
    n_estimators=500,
    subsample_for_bin=500,
    objective='regression',
    class_weight=None,
    min_split_gain=0,
    min_child_weight=0.001,
    min_child_samples=10,
    subsample=0.8,
    subsample_freq=30,
    colsample_bytree=0.9,
    colsample_bynode=0.1,
    reg_alpha=0.1,
    reg_lambda=0.05,
    random_state=None,
    n_jobs=-1,
    silent=True,
    importance_type='split'
)

lgbm = lgb_reg.fit(X=X_train, y=y_train,
                  eval_set = eval_set, verbose=0)

y_pred_train = lgbm.predict(X_train)
y_pred_val = lgbm.predict(X_val)

train_rmse = mean_squared_error(y_true=y_train, y_pred=y_pred_train, squared=False)
val_rmse = mean_squared_error(y_true=y_val, y_pred=y_pred_val, squared=False)
print(f'RMSE Train: {train_rmse: .8f}')
print(f'RMSE Valid: {val_rmse: .8f}\tRatio (val/train): {val_rmse/train_rmse: .5f}')

plot_lgb_history(lgbm)
```
