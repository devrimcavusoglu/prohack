```python
import h2o
from h2o.frame import H2OFrame
from h2o.automl import H2OAutoML; h2o.init()
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
```

```python
def plot_lgb_history(history):
    for i, key in enumerate(history.evals_result_):
        if i == 0:
            res = pd.DataFrame(history.evals_result_[key])
            res['index'] = np.arange(len(history.evals_result_[key]['l2']))
            res['type'] = key
        temp = pd.DataFrame(history.evals_result_[key])
        temp['type'] = key
        temp['index'] = np.arange(len(history.evals_result_[key]['l2']))
        
        res = res.append(temp, ignore_index=True)
        del temp; gc.collect()
    
    fig = px.scatter(data_frame=res, x='index', y='l2', color='type',
                 title='Learning curve')
    fig.show()
```

```python
train = pd.get_dummies(train)
```

```python
scaler = MinMaxScaler()

train_temp = train.drop(['galactic year', 'y'], axis=1)
train_sc = scaler.fit_transform(train_temp)
train_sc = pd.DataFrame(train_sc, columns=train_temp.columns)
del train_temp; gc.collect()

train_sc['galactic year'] = train['galactic year']
train_sc['y'] = train['y']
train_sc
```

```python
train.rename(columns={
    "Population, ages 15â€“64 (millions)": "pop_age_15_64_mil",
    "Unemployment, youth (% ages 15â€“24)": "unemp_youth_perc_ages_15_24",
    "galaxy_BoÃ¶tes I": "galaxy_Bootes_i",
    "galaxy_BoÃ¶tes II": "galaxy_Bootes_ii",
    "galaxy_BoÃ¶tes III": "galaxy_Bootes_iii",
    "galaxy_BoÃ¶tes IV": "galaxy_Bootes_iv"
}, inplace=True)
```

```python
last_n_years = 3
val_year = train['galactic year'].unique()[-last_n_years]
train_set = train[train['galactic year']<val_year]
val_set = train[train['galactic year']>=val_year]
```

```python
y = 'y'
X = train_set.columns.drop(y).tolist()
```

```python
train_set = H2OFrame(train_set)
val_set = H2OFrame(val_set)

y_train = train_set[y].as_data_frame().values
y_val = val_set[y].as_data_frame().values

# No need for eval set
```

```python
# num_leaves << 2^(max_depth)

aml = H2OAutoML(
    nfolds=0, # if nfolds != 0, then validation_frame in train() is ignored.
    balance_classes=False,
    class_sampling_factors=None,
    max_after_balance_size=5.0,
    max_runtime_secs=300,
    max_runtime_secs_per_model=None,
    max_models=None,
    stopping_metric='AUTO',
    stopping_tolerance=None,
    stopping_rounds=3,
    seed=None,
    project_name=None,
    exclude_algos=None,
    include_algos=None,
    exploitation_ratio=0,
    modeling_plan=None,
    monotone_constraints=None,
    algo_parameters=None,
    keep_cross_validation_predictions=False,
    keep_cross_validation_models=False,
    keep_cross_validation_fold_assignment=False,
    sort_metric='RMSE',
    export_checkpoints_dir=None,
    verbosity='warn'
)

aml.train(x=X, y=y,
                training_frame=train_set,
                validation_frame=val_set
    )

y_pred_train = aml.predict(train_set[X]).as_data_frame().values
y_pred_val = aml.predict(val_set[X]).as_data_frame().values

train_rmse = mean_squared_error(y_true=y_train, y_pred=y_pred_train, squared=False)
val_rmse = mean_squared_error(y_true=y_val, y_pred=y_pred_val, squared=False)
print('#'*60)
print(f'RMSE Train: {train_rmse: .8f}')
print(f'RMSE Valid: {val_rmse: .8f}\tRatio (val/train): {val_rmse/train_rmse: .5f}')
```
