```python
test = pd.read_csv('./outputs/imputed_test_mfc2.csv')                
m_test, n_test = test.shape                
print('Data Frame: {} x {}'.format(m_test,n_test))                
test.head()
```

```python
df = train.append(test, ignore_index=True)
df
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, MaxAbsScaler
```

```python
df_scaled = df.drop(['galactic year', 'galaxy', 'y'], axis=1)
cols = df_scaled.columns
```

```python
scaler = StandardScaler()
# scaler = MinMaxScaler()
# scaler = MaxAbsScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df_scaled), columns=cols)
df_scaled
```

```python
df_scaled['galaxy'] = df['galaxy']
df_scaled = pd.get_dummies(df_scaled, columns=['galaxy'])
df_scaled
```

```python
from sklearn.model_selection import train_test_split
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model
```

```python
ncol = df_scaled.shape[1]
encoding_dim = 50
```

```python
input_dim = Input(shape = (ncol, ))

# Encoder Layers
encoded1 = Dense(200, activation = 'relu')(input_dim)
encoded2 = Dense(100, activation = 'relu')(encoded1)
encoded3 = Dense(75, activation = 'relu')(encoded2)
encoded4 = Dense(encoding_dim, activation = 'relu')(encoded3)

# Decoder Layers
decoded1 = Dense(75, activation = 'relu')(encoded4)
decoded2 = Dense(100, activation = 'relu')(decoded1)
decoded3 = Dense(200, activation = 'relu')(decoded2)
decoded5 = Dense(ncol, activation = 'sigmoid')(decoded3)

# Combine Encoder and Deocder layers
autoencoder = Model(inputs = input_dim, outputs = decoded5)

# Compile the Model
autoencoder.compile(optimizer = 'adadelta', loss = 'binary_crossentropy')
autoencoder.summary()
```

```python
X_train = df_scaled.to_numpy()
```

```python
history = autoencoder.fit(X_train, X_train, epochs = 20, batch_size = 32, shuffle = False, validation_data = (X_train, X_train))
```

```python
encoder = Model(inputs = input_dim, outputs = encoded4)
encoded_input = Input(shape = (encoding_dim, ))
```

```python
encoded_df = pd.DataFrame(encoder.predict(X_train))
encoded_df = encoded_df.add_prefix('feature_')
encoded_df.head(m_train+m_test-5)
```
